```python
import re
import json
```

```python
try:
    with open("projectLinks_ban.txt", "r") as f:
        projectLinks_ban = [i for i in f.read().split('\n')]
        f.close()
except:
    projectLinks_ban = []
    
try:
    with open("projectLinks_thue.txt", "r") as f:
        projectLinks_thue = [i for i in f.read().split('\n')]
        f.close()
except:
    projectLinks_thue = []

projectLinks = projectLinks_thue + projectLinks_ban
new_projectLinks = list(dict.fromkeys(projectLinks))

with open("ProjectInfoInsertCommand.txt", "w", encoding='utf-8') as f:
    f.write("")
    f.close()

with open("ProjectInfoInsertCommand.txt", "a") as f:
    for projectlink in new_projectLinks:
        f.write("('{}')".format(projectlink.replace("https://", ""))+",\n")
    f.close()
```
